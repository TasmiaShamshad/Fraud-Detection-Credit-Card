# Exploratory Data Analysis to Understand the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , f1_score
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv(r"C:\Users\HP\Downloads\creditcard.csv")
df.head() # to check first 500 rows
df.info() 
df.isnull().sum() # to check if there is anyother missing value
df['Class'].value_counts() # to check the fradulent and normal cases, binary classifiaction
```

### Grouping of two classes and Comparison

```python
legit = df[df.Class == 0]
fraud = df[df.Class ==1]
print(legit.shape)
print(fraud.shape)
a= legit.Amount.describe()
b= fraud.Amount.describe()
print(a)
print(b)
df.groupby('Class').mean()
```

#### Undersampling of majority dataset

```python
# We will build a new dataset of almost same numbers of legit transactions to train the model /
# better either you increase the minority by SMOTE or decrease the dataset to build a pattern. 
```

```python
Under_sample = legit.sample(1000) # random sampling 
new_dataset = pd.concat([Under_sample, fraud], axis = 0) # 0 is rows (in row format)and column is 1
new_dataset.tail()
```

## Splitting Data

```python
## Removing the class column and so to train the model next
X = new_dataset.drop(columns= 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state= 2, stratify= Y )
```

### Model Training

```python
# training with teh training data with logistic Regression, for binary classifications(0,1 outputs)
model = LogisticRegression()
model.fit(X_train, Y_train)
```

### Model Evaluation

```python
## Model Evaluation with Accuracy Score
X_train_prediction = model.predict(X_train)
Accuracy_Y_train = accuracy_score(X_train_prediction, Y_train)
print("Accuracy score for X_train is, ", Accuracy_Y_train)

from sklearn.metrics import f1_score
print(" ")
## Model Evaluation by F1 score
X_train_prediction = model.predict(X_train)
Accuracy_Y_train = accuracy_score(X_train_prediction, Y_train)
F1_Y_train = f1_score(Y_train, X_train_prediction)

print("F1 Score on Training Data:", F1_Y_train)
```

## Command-line Interface

```python
print("Welcome to the Fraud Detection System (Index-Based Checker)")
print(f"Total records available: {len(new_dataset)}")

while True:
    try:
        index = int(input("Enter the index number of the transaction to check (or No to exit): "))
        if index == 'No':
            print("Program is closed, We are Exiting...")
            break

        if index < 0 or index >= len(new_dataset):
            print("❌ Invalid index. Please enter a number between 0 and", len(new_dataset)-1)
            continue

        # Get features and actual class
        input_data = X.iloc[index].values.reshape(1, -1)
        actual_class = Y.iloc[index]

        # Predict
        prediction = model.predict(input_data)[0]
        prediction_proba = model.predict_proba(input_data)[0][1]

        print(f"\nTransaction at index {index}:")
        print(f"Model Prediction: {'⚠️ FRAUD' if prediction == 1 else '✅ Not Fraud'}")
        print(f"Actual Class:     {'⚠️ FRAUD' if actual_class == 1 else '✅ Not Fraud'}")
        print(f"Fraud Probability: {prediction_proba:.2f}\n")

    except ValueError:
        print("⚠️ Please enter a valid number.")
```
